Spliting of data as input and output 

```python
x=df.drop(["Income"],axis=1)
y=df["Income"]
print(x.shape,y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_auc_score
```

```python
def metrics(a,b):
    print("accuracy score : ",accuracy_score(a,b))
    print("confusion matrix : ",confusion_matrix(a,b))
    print("classification_report : ",classification_report(a,b))
    print("roc_auc_score : ",roc_auc_score(a,b))
```

```python
max_acc=0
best_rand=0
for i in range(200):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LogisticRegression()
    lr.fit(xtrain,ytrain)
    pred=lr.predict(xtest)
    acc=accuracy_score(ytest,pred)
    if acc>max_acc:
        max_acc=acc
        best_rand=i
print("max_acc = ",max_acc,"best rand = ",best_rand)
```

```python
def model(mod,x1,y1,x2,y2):
    mod.fit(x1,y1)
    pred=mod.predict(x2)
    metrics(y2,pred)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.30,random_state=56)
```

```python
from sklearn.model_selection import cross_val_score
print("cross_val_score : ",cross_val_score(LogisticRegression(),x,y,cv=4).mean())
model(LogisticRegression(),xtrain,ytrain,xtest,ytest)
```

```python
print("cross_val_score : ",cross_val_score(GaussianNB(),x,y,cv=4).mean())
model(GaussianNB(),xtrain,ytrain,xtest,ytest)
```

```python
print("cross_val_score : ",cross_val_score(DecisionTreeClassifier(),x,y,cv=4).mean())
model(DecisionTreeClassifier(),xtrain,ytrain,xtest,ytest)
```

```python
print("cross_val_score : ",cross_val_score(KNeighborsClassifier(),x,y,cv=4).mean())
model(KNeighborsClassifier(),xtrain,ytrain,xtest,ytest)
```

```python
print("cross_val_score : ",cross_val_score(RandomForestClassifier(),x,y,cv=4).mean())
model(RandomForestClassifier(),xtrain,ytrain,xtest,ytest)
```

```python
print("cross_val_score : ",cross_val_score(AdaBoostClassifier(),x,y,cv=4).mean())
model(AdaBoostClassifier(),xtrain,ytrain,xtest,ytest)
```

RandomForestClassifier is looks to best with high accuracy when compare with others

Hyper Parameter Tuneing

```python
from sklearn.model_selection import GridSearchCV
para={"criterion":["gini","entropy"],}
gv=GridSearchCV(RandomForestClassifier(),para,cv=5)
9gv.fit(xtrain,ytrain)
gvpred=gv.predict(xtest)
accuracy_score(gvpred,ytest)
```
